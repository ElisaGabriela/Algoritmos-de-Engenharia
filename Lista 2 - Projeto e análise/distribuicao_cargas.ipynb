```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def distribuir_tarefas(cargas, k):
    n = len(cargas)
    tarefas_por_trabalhador = n // k
    trabalhadores = [[] for _ in range(k)]
    
    for i in range(n):
        trabalhador_atual = i // tarefas_por_trabalhador
        if trabalhador_atual < k:  
            trabalhadores[trabalhador_atual].append(cargas[i])
    
    cargas_trabalhadores = [sum(tarefas) for tarefas in trabalhadores]
    
    return max(cargas_trabalhadores), cargas_trabalhadores
```

```python
np.random.seed(42)
```

```python
def vetor_ordenado(n):
    return list(range(n + 1))
```

```python
n = 100000000  
k = 10   
cargas = vetor_ordenado(n)
```

```python
max_carga_sem_permutacao, cargas_sem_permutacao = distribuir_tarefas(cargas, k)
```

```python
random.shuffle(cargas)
max_carga_com_permutacao, cargas_com_permutacao = distribuir_tarefas(cargas, k)
```

```python
print(f"Máxima carga sem permutação: {max_carga_sem_permutacao}")
print(f"Máxima carga com permutação: {max_carga_com_permutacao}")
```

```python
trabalhadores = list(range(1, k + 1))

plt.figure(figsize=(10, 6))

plt.plot(trabalhadores, cargas_sem_permutacao, label="Sem permutação", marker='o', color='b')

plt.plot(trabalhadores, cargas_com_permutacao, label="Com permutação", marker='x', color='r')

plt.title("Comparação de distribuição de cargas entre trabalhadores")
plt.xlabel("Trabalhadores")
plt.ylabel("Carga Total")
plt.legend()

plt.grid(True)
plt.show()
```
