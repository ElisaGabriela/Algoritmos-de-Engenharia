```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np
import math
from random import randint
from math import ceil
```

```python
np.random.seed(42)
```

```python
def median(coordinates):
    sorted_coords = sorted(coordinates)
    n = len(sorted_coords)
    
    if n % 2 == 1:
        return sorted_coords[n // 2]
    else:
        return (sorted_coords[(n // 2) - 1] + sorted_coords[n // 2]) / 2
```

```python
def find_best_location(points):
    x_coords = [point[0] for point in points]
    y_coords = [point[1] for point in points]

    best_x = median(x_coords)
    best_y = median(y_coords)

    return (best_x, best_y)
```

```python
num_points = 10
points = [(random.randint(0, 100), random.randint(0, 100)) for _ in range(num_points)]
```

```python
optimal_location = find_best_location(points)
```

```python
x_coords, y_coords = zip(*points) 
plt.figure(figsize=(10, 8))

plt.scatter(x_coords, y_coords, color="blue", label="Casas")

plt.scatter(*optimal_location, color="red", marker="x", s=100, label="Agência")

plt.xlabel("Coordenada x")
plt.ylabel("Coordenada y")
plt.title("Localização ótima para o problema da agência de correios")
plt.legend()
plt.grid()

plt.show()
```

```python
np.random.seed(42)  
num_points = 1000
x_coords = np.random.randint(0, 100, num_points)
y_coords = np.random.randint(0, 100, num_points)
weights = np.random.randint(1, 5, num_points)
```

```python
def weighted_median(values, weights):
    sorted_indices = np.argsort(values)
    values_sorted = values[sorted_indices]
    weights_sorted = weights[sorted_indices]
    cumulative_weights = np.cumsum(weights_sorted)
    cutoff = cumulative_weights[-1] / 2
    median_index = np.where(cumulative_weights >= cutoff)[0][0]
    return values_sorted[median_index]
```

```python
optimal_x = weighted_median(x_coords, weights)
optimal_y = weighted_median(y_coords, weights)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x_coords, y_coords, s=weights * 30, c='blue', label='Casas')
plt.scatter(optimal_x, optimal_y, s=100, c='red', marker='X', label='Posto dos Correios')
#for i in range(num_points):
#    plt.text(x_coords[i] + 1, y_coords[i] + 1, f'{weights[i]}', fontsize=10)

plt.xlabel('X Coordenada')
plt.ylabel('Y Coordenada')
plt.title('Melhor localização para o posto dos correios')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def insertion_sort(valores, pesos):
    for i in range(1, len(valores)):
        k_valor = valores[i]
        k_peso = pesos[i]
        j = i - 1
        while j >= 0 and valores[j] > k_valor :
            valores[j + 1] = valores[j]
            pesos[j + 1] = pesos[j]
            j -= 1
        valores[j + 1] = k_valor 
        pesos[j + 1] = k_peso 
```

```python
def mediana_ponderada(valores, pesos):
    insertion_sort(valores, pesos)
    
    soma_pesos = sum(pesos)
    
    peso_acumulado = 0
    for valor, peso in zip(valores, pesos):
        peso_acumulado += peso 
        if peso_acumulado >= soma_pesos / 2:
            return valor
```
